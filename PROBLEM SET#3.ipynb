```python
Question 1
Introduction:
Special thanks to: https://github.com/justmarkham for sharing the dataset and 
materials.
Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men
```

```python
# IMPORTING THE NECESSARY LIBRARIES
import pandas as pd  

# IMPORTING THE DATASET AND ADDING "users" VARIABLE TO IMPORT THE DATASET
URL = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(URL, sep='|') 
MEAN_AGE = users.groupby('occupation')['age'].mean()  # CALCULATING MEAN AGE PER OCCUPATION 

# CALCULATING THE MALES RATIO PER OCCUPATION AND SORT IT FROM MOST TO THE LEAST

MALE_RATIO = users[users['gender'] == 'M'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()
MALE_RATIO = MALE_RATIO.sort_values(ascending=False)

# CALCULATING THE MINIMUM AND MAXIMUM AGE FOR EACH OCCUPATION

MIN_MAXAGE = users.groupby('occupation')['age'].agg(['min', 'max'])

# CALCULATING THE MEAN AGE FOR EACH COMBINATION OF OCCUPATION AND GENDER
MEAN_AGE_GENDER = users.groupby(['occupation', 'gender'])['age'].mean()

# DISPLAY THE PERCENTAGE OF WOMEN AND MEN FOR EACH OCCUPATION 
GENDER_OCCUPATION = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
OCCUPATION_COUNT = users.groupby(['occupation']).agg('count')
GENDER_OCCUPATION_RATIO = GENDER_OCCUPATION.div(OCCUPATION_COUNT, level = "occupation") * 100
```

```python
from IPython.display import display

# THUS IT WILL BE PRINTING THE MEAN AGE, MALE RATIO, MIN MAX AGE, MEAN AGE BY GENDER AND GENDER OCCUPATION RATIO
print("The mean age per occupation: ")
display(MEAN_AGE)
print("The Male ratio per occupation:")
display(MALE_RATIO)
print("The minimum and maximum ages for each occupation:")
display(MIN_MAXAGE)
print("The mean age for each combination of occupation and sex:")
display(MEAN_AGE_GENDER)
print("The percentage of women and men for each occupation:")
display(GENDER_OCCUPATION_RATIO)
```

```python
# IMPORTING THE NECESSARY LIBRARIES
import pandas as pd

# IMPORTING THE DATASET FROM URL
DATA_FROM_URL = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"  # Replace with the actual URL of the dataset
euro12 = pd.read_csv(DATA_FROM_URL)

# ASSIGNING THE VARIABLE NAMED euro12

# SELECTING THE GOALS COLUMN
goals_column = euro12['Goals']

# CALCULATING THE NUMBER OF TEAMS PARTICIPATED
num_teams_participated = euro12['Team'].nunique()

# CALCULATING THE NUMBER OF COLUMNS IN THE DATASET
num_columns = euro12.shape[1]

# PRESENTING THE COLUMNS FOR TEAM, YELLOW CARDS, AND RED CARDS AND ASSIGN THE DATAFRAME NAMED 'discipline'
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]

# SORTING THE TEAMS BY RED CARDS AND THEN BY YELLOW CARDS
discipline = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[True, True])

# CALCULATING THE MEAN OF YELLOW CARDS PER TEAM
mean_yellow_cards_per_team = discipline.groupby('Team')['Yellow Cards'].mean()

# FILTERING THE TEAMS THAT SCORED MORE THAN >GOALS
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]

# SELECTING TEAM STARTS WITH G
teams_start_with_G = euro12[euro12['Team'].str.startswith('G')]

# SELECTING 7 COLUMNS
first_7_columns = euro12.iloc[:, :7]

# SELECTING ALL COLUMNS EXCEPT LAST 3
all_columns_except_last_3 = euro12.iloc[:, :-3]

# SHOWING THE SHOOTING ACCURACY ENGLAND, ITALY AND RUSSIA
selected_teams = ['England', 'Italy', 'Russia']
ShootingAccuracy_Selected_Teams = euro12.loc[euro12['Team'].isin(selected_teams), ['Team', 'Shooting Accuracy']]

# PRINTING THE RESULTS
print("\nNumber of teams participated in Euro2012:", num_teams_participated)
print("\nNumber of columns in the dataset:", num_columns)
print("\nDataFrame 'discipline':")
print(discipline)
print("\nMean Yellow Cards per Team:")
print(mean_yellow_cards_per_team)
print("\nTeams that scored more than 6 goals:")
print(teams_more_than_6_goals)
print("\nTeams that start with 'G':")
print(teams_start_with_G)
print("\nFirst 7 columns:")
print(first_7_columns)
print("\nAll columns except the last 3:")
print(all_columns_except_last_3)
print("\nShooting Accuracy of England, Italy, and Russia:")
print(ShootingAccuracy_Selected_Teams)
```

```python
# IMPORTING DATATIME AND FROM datetime IMPORTING date
from datetime import date
import datetime as dt 

# READING THE DATA FROM wind.txt, ASSIGNING THE data AND PARSING DATES 
data= pd.read_fwf('wind.txt',parse_dates=[['Yr','Mo','Dy']])
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data.head()

data["Yr_Mo_Dy"] = np.where(pd.DatetimeIndex(data["Yr_Mo_Dy"]).year < 2061,data.Yr_Mo_Dy,data.Yr_Mo_Dy - pd.offsets.DateOffset(years=100))
data.head()


# SETTING THE RIGHT DATES AS THE INDEX AND DISPLAY THE DATATYPES
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
display(data.dtypes)  # DISPLAYING THE DATA TYPES 
data = data.set_index('Yr_Mo_Dy')
data.index
data  # SHOWING THE DATA AFTER INDEXING

# CALCULATING MISSING VALUES
missing_values = data.isnull().values.sum()
new_data = data.dropna()
# CALCULATING NON-MISSING VALUES
nonmissing_values = data.notnull().sum()
# CALCULATING THE MEAN WINDSPEEDS
mean = data.mean().mean()


# CALCULATING THE MINIMUM, MAXIMUM, MEAN AND STANDARD DEVIATION AT EACH LOCATION 
mini=new_data.min()
maxi=new_data.max()
mean=new_data.mean()
std=new_data.std()
var=[mini,maxi,mean,std]
index=["Min","Max","Mean","Std"]
loc_stats= pd.DataFrame(var,index)
loc_stats
# CALCULATE THE MAX, MIN, AND MEAN WINDSPEED AND STANDARD DEVIATION OF WINDSPEEDS ACROSS ALL THE LOCATIONS
day_stats = pd.concat([new_data.min(axis=1), new_data.max(axis=1), new_data.mean(axis=1), new_data.std(axis=1)], axis=1)
day_stats.rename(columns={0:'Min',1:'Max',2:'Mean',3:'Std'}, inplace=True)
day_stats

new_data[new_data.index.month==1].mean()

downsample_data = data.resample('Y').ffill()
downsample_data

month_downsample= data.resample('M').ffill()
month_downsample.head()

week_downsample= data.resample('W').ffill()
week_downsample.head()

week_stat=week_downsample.groupby(week_downsample.index.to_period('W')).agg(['min','max','mean','std'])
week_stat.loc[week_stat.index[1:53],:].head()

```

```python
# Filter the data for the specific years (1900, 1950, and 2000)
years_of_interest = [1900, 1950, 2000]
data_of_interest = data[data['Year'].isin(years_of_interest)]

# Extract data from the filtered DataFrame
years = data_of_interest['Year']
marriages_per_capita = data_of_interest['Marriages'] / data_of_interest['Population']
divorces_per_capita = data_of_interest['Divorces'] / data_of_interest['Population']

# Create the vertical bar chart
plt.figure(figsize=(10, 8))
bar_width = 10
index = years - bar_width / 2

plt.bar(index, marriages_per_capita, width=bar_width, label='Marriages per Capita')
plt.bar(index + bar_width, divorces_per_capita, width=bar_width, label='Divorces per Capita')

# Label the axes and add a title
plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')

# Set the xticks and xtick labels
plt.xticks(index + bar_width / 2, years)

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python

# COUNTING THE NUMBER OF ASSASSINATED EMPERORS
assassinated_emperors = data[data['Cause_of_Death'] == 'Assassinated']['Emperor'].count()

# COUNTING THE NUMBER OF NON_ASSASSINATED EMPERORS
non_assassinated_emperors = data[data['Cause_of_Death'] != 'Assassinated']['Emperor'].count()

# CREATING THE PIE CHART
labels = ['Assassinated Emperors', 'Non-Assassinated Emperors']
sizes = [assassinated_emperors, non_assassinated_emperors]
colors = ['brown', 'grey']
explode = (0.1, 0)  

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=100)

# ADDING A CIRCLE at the center to make it an even circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# ADDING THE TITLE
plt.title('Fraction of Roman Emperors Assassinated', fontsize=16, fontweight='bold', color='black')

# EQUAL ASPECT RATIO ENSURES THAT PIE IS DRAWN AS A CIRCLE
plt.axis('equal')

# SHOWING THE PLOT
plt.show()   
```
